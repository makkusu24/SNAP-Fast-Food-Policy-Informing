```
snap = _deepnote_execute_sql('SELECT *\nFROM \'snap_deserts.csv\' AS snap', 'SQL_DEEPNOTE_DATAFRAME_SQL', audit_sql_comment='', sql_cache_mode='cache_disabled')
snap
```

```
city_to_county_mapping = _deepnote_execute_sql('SELECT city, state_id, county_name\nFROM \'uscities.csv\' AS city_to_county_mapping', 'SQL_DEEPNOTE_DATAFRAME_SQL', audit_sql_comment='', sql_cache_mode='cache_disabled')
city_to_county_mapping
```

```
#fast_food.columns
"""
As we can see, both dataframes have vast amounts of data. There's a high chance of data 
entry errors so we should impute any possible missing values. To save on computation, .isnull().sum()
shows that NaN values do not exceed even 0.05% of the data, so we can impute with 0 rather than means, as it
will have trivial impact on results.
"""
fast_food = fast_food.fillna(0)
snap = snap.fillna(0)
fast_food.shape, snap.shape
```

```
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
import geopandas as gpd 
```

```
"""
Data processing stage:
"""
fast_food_grouped = fast_food.groupby('province').size().reset_index(name='FastFoodCount')

fast_food_grouped.head(10)
```

```
fast_food['City_State_Identifier'] = fast_food['city'] + '_' + fast_food['province']
city_to_county_mapping['City_State_Identifier'] = city_to_county_mapping['city'] + '_' + city_to_county_mapping['state_id']

# Merge the fast-food data with the mapping data to get the corresponding 'County' for each entry
fast_food_with_county = pd.merge(fast_food, city_to_county_mapping[['City_State_Identifier', 'county_name']], 
                                 on='City_State_Identifier', how='left')

# Rename the 'county_name' column to 'County'
fast_food_with_county.rename(columns={'county_name': 'County'}, inplace=True)

# Drop the temporary 'City_State_Identifier' columns
fast_food_with_county.drop('City_State_Identifier', axis=1, inplace=True)
city_to_county_mapping.drop('City_State_Identifier', axis=1, inplace=True)

# Show the first few rows of the fast-food data with the new 'County' column
fast_food_with_county.loc[:, ['city', 'name', 'province', 'County']].head(10)
```

```
fast_food_grouped_county = fast_food_with_county.groupby(['County', 'province']).size().reset_index(name='FastFoodCount')

# Merge the newly grouped fast-food data with the SNAP data, which is already grouped by 'County' and 'State_Abbreviation'
merged_data_county = pd.merge(snap, fast_food_grouped_county, 
                              left_on=['County', 'State_Abbreviation'], 
                              right_on=['County', 'province'], 
                              how='inner')

# Calculate the density of fast-food restaurants per capita and the proportion of SNAP recipients
merged_data_county['FastFoodDensity'] = merged_data_county['FastFoodCount'] / merged_data_county['POP2010']
merged_data_county['SNAPProportion'] = merged_data_county['TractSNAP'] / merged_data_county['POP2010']

# Calculate the correlation between FastFoodDensity and SNAPProportion at the county level
correlation_matrix_county = merged_data_county[['FastFoodDensity', 'SNAPProportion']].corr()

# Show the correlation matrix and the first few rows of the merged data at the county level
merged_data_county.head(10)
```

```
correlation_matrix_county
```

```
"""
High SNAP Proportion: These counties have a higher proportion of SNAP recipients, 
which could indicate a higher level of food insecurity.
Low Fast Food Density: At the same time, these counties have a lower density of 
fast-food restaurants, suggesting fewer quick and cheap food options.
"""
# Calculate the median values for FastFoodDensity and SNAPProportion for further analysis
median_fast_food_density_county = merged_data_county['FastFoodDensity'].median()
median_snap_proportion_county = merged_data_county['SNAPProportion'].median()

# Identify counties with above-median SNAPProportion and below-median FastFoodDensity
target_counties = merged_data_county[(merged_data_county['FastFoodDensity'] > median_fast_food_density_county) & 
                                     (merged_data_county['SNAPProportion'] > median_snap_proportion_county)]

# Group these target counties by State and County to see where interventions could be most impactful
target_counties_grouped = target_counties.groupby(['State', 'County']).size().reset_index(name='NumOfTargetTracts')

# Sort the grouped data by the number of target tracts in descending order to prioritize interventions
target_counties_grouped = target_counties_grouped.sort_values(by='NumOfTargetTracts', ascending=False)

# Show the top 10 target counties for potential interventions
target_counties_grouped.head(10)
```

```
top_counties = target_counties_grouped.head(50)
heatmap_data = merged_data_county.groupby(['State', 'County']).agg({
    'FastFoodDensity': 'mean',
    'SNAPProportion': 'mean'
}).reset_index()
filtered_heatmap_data = heatmap_data[heatmap_data['County'].isin(top_counties['County']) & heatmap_data['State'].isin(top_counties['State'])]

# Create the heatmap for these top 50 counties
plt.figure(figsize=(12, 8))
filtered_heatmap_pivot = filtered_heatmap_data.pivot("County", "State", "FastFoodDensity")
sns.heatmap(filtered_heatmap_pivot, cmap="coolwarm", annot=True, fmt=".2f", cbar=True)
plt.title('Heatmap of Average Fast-Food Density for Top 50 High-SNAP, High-Density Counties')
plt.xlabel('State')
plt.ylabel('County')
plt.show()
```

```
plt.figure(figsize=(12, 8))
sns.scatterplot(x='FastFoodDensity', y='SNAPProportion', hue='State', data=merged_data_county, palette='tab20', s=100, alpha=0.6)
plt.title('Scatter Plot of Fast-Food Density vs. SNAP Proportion by County')
plt.xlabel('Fast-Food Density (Per Capita)')
plt.ylabel('SNAP Proportion (Per Capita)')
plt.legend(title='State', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
